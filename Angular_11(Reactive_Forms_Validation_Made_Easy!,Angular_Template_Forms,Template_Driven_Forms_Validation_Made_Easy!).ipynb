 **Form validation means checking user input in a form to make sure the data entered is correct, complete, and acceptable before it is submitted or processed.**

# ** In Simple Terms:**
Form validation ensures:

Required fields are not left empty.

Email addresses are in a valid format.

Passwords meet strength rules (e.g., minimum length).

Numbers fall within allowed ranges.



# **app.html**



```
<h1>validation in Reactive Forms</h1>

<form [formGroup]="profileForm" (ngSubmit)="submitData()">
  
  <input type="text" placeholder="enter name" formControlName="name">
  <span *ngIf="name?.hasError('required')">This is Required field</span>
  <span class="error" *ngIf="name?.invalid && (name?.dirty || name?.touched)">
    Please enter valid input
  </span>
  <br><br>

  <input type="text" placeholder="enter password" formControlName="password">
  <span *ngIf="password?.hasError('required')">This is Required field</span>
  <span class="error" *ngIf="password?.invalid && (password?.dirty || password?.touched)">
    Please enter valid password
  </span>
  <br><br>

  <input type="text" placeholder="enter email" formControlName="email">
  <span *ngIf="email?.hasError('required')">This is Required field</span>
  <span class="error" *ngIf="email?.invalid && (email?.dirty || email?.touched)">
    Please enter valid email
  </span>
  <br><br>

  <button>Submit</button>
</form>

```



# **app.ts**



```
import { Component } from '@angular/core';
import { FormControl, FormGroup, Validators, ReactiveFormsModule } from '@angular/forms';
import { NgIf } from '@angular/common';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [ReactiveFormsModule, NgIf],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  profileForm = new FormGroup({
    name: new FormControl('', [Validators.required]),
    password: new FormControl('', [Validators.required, Validators.minLength(5)]),
   
    email: new FormControl('', [Validators.required,Validators.maxLength(30),Validators.pattern(/^[a-zA-Z]{1}[a-zA-Z0-9._%+-]*@[a-zA-Z.-]+\.[a-zA-Z]{2,}$/)
])

  });

  submitData() {
    console.log(this.profileForm.value);
  }

  get name() {
    return this.profileForm.get('name');
  }

  get password() {
    return this.profileForm.get('password');
  }

  get email() {
    return this.profileForm.get('email');
  }
}

```

```
<h1>Template Driven Forms</h1>

<form #userForm="ngForm" (ngSubmit)="addUser(userForm.value)">
  <!-- Name Field -->
  <input
    #name="ngModel"
    type="text"
    name="name"
    ngModel
    required
    placeholder="enter name"
  />
  <span class="error" *ngIf="name.invalid && name.touched">
    Name is required
  </span>

  <br /><br />

  <!-- Password Field -->
  <input
    #password="ngModel"
    type="password"
    name="password"
    ngModel
    placeholder="enter password"
    minlength="5"
    maxlength="10"
    required
  />
  <span class="error" *ngIf="password.invalid && password.touched">
    Please enter valid Password
  </span>

  <br /><br />

  <!-- Email Field -->
  <input
    #email="ngModel"
    type="email"
    name="email"
    ngModel
    placeholder="enter email"
    required
  />
  <span class="error" *ngIf="email.invalid && email.touched">
    Please enter valid email
  </span>

  <br /><br />

  <!-- Address Field -->
  <input
    #address="ngModel"
    type="text"
    name="address"
    ngModel
    placeholder="enter address"
    pattern="[a-zA-Z0-9\s]+"
    required
  />
  <span class="error" *ngIf="address.invalid && address.touched">
    Please enter valid address
  </span>

  <br /><br />

  <!-- Gender Select -->
  <select name="gender" ngModel required>
    <option value="">Select Gender</option>
    <option value="male">Male</option>
    <option value="female">Female</option>
  </select>

  <br /><br />

  <!-- Submit Button -->
  <button [disabled]="userForm.invalid">Add User</button>
</form>

```
