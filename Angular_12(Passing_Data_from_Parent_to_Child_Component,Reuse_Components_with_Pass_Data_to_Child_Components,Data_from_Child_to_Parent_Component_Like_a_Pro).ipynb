# **“Pass data Child to Parent Component” outlines the process of sending data from a child component back to a parent component in Angular.**

# **app.ts**



```
import { Component } from '@angular/core';
import { CommonModule } from '@angular/common';
import { FormsModule } from '@angular/forms';

import { User } from './user/user'; // Ensure UserComponent is exported as standalone

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [CommonModule, FormsModule, User],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  users: string[] | undefined;

  handleUsers(Users: string[]) {
    console.log(Users); // ✅ Corrected the log
    this.users = Users;
  }
}

```



# **app.html**



```
<h1>pass data to child component</h1>
<app-user (getUsers)="handleUsers($event)"></app-user>

@for(user of users;track user){
    <h1>{{user}}</h1>
}
```



# **user.ts**



```
import { Component, EventEmitter, Output } from '@angular/core';

@Component({
  selector: 'app-user',
  standalone: true,
  templateUrl: './user.html',
  styleUrls: ['./user.css'],
  imports: []
})
export class User {
  @Output() getUsers = new EventEmitter<string[]>();

  users = ['Anil', 'Sidhu', 'Sam', 'Peter', 'Bruce'];

  ngOnInit() {
    this.loadData();
  }

  loadData() {
    this.getUsers.emit(this.users);
  }
}

```



# **user.html**



```
<p>User Component</p>

<!-- This calls the loadData() method correctly -->
<button (click)="loadData()">Pass Data 1</button>

```
