

# **app.ts**



```
import { Component, signal } from '@angular/core';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  count = signal(10); // ✅ Declare the signal

  updateValue(val: string) {
    if (val === 'dec') {
      this.count.set(this.count() - 1);
    } else {
      this.count.set(this.count() + 1);
    }
  }
}
```

# **Data Type with Signals**

# **1️⃣ Define Signals**



```
import { signal } from '@angular/core';

const count = signal(0); // default number signal

```



#**2️⃣ Define Data Type for Signals**



```
const name = signal<string>("Ajay");
const isOnline = signal<boolean>(true);
const age = signal<number>(25);

```



# **3️⃣ Define Data Type for Values (inside a signal)**



```
const user = signal<{ id: number, name: string }>({ id: 1, name: "Ajay" });

```



# **4️⃣ Update Signals**
✅ set() method – Directly replace the value



```
count.set(5);

```



# **⚠️ update() method – Used to transform the current value**



```
count.update(val => val + 1);  // Limited use because it only works with simple mutations

```



# **app.html:**



```
<h1>Types with Signal</h1>
<h1>{{ data() }}</h1>

<button (click)="updateValue()">Update</button>

```



# **app.ts**



```
import { Component, signal, computed, WritableSignal, Signal } from '@angular/core';

import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  // A writable signal of number type
  data: WritableSignal<number> = signal(10);

  // A computed signal that depends on 'data'
  //count: Signal<number> = computed(() => this.data() * 2);

  updateValue() {
    // Update current signal value by adding 10
    this.data.update(val => val + 10);
  }
}
```
