# ** 1. What is a Directive in Angular?**
A directive is a class in Angular that adds behavior to elements in your Angular application.
It allows you to manipulate the DOM structure, style, or behavior of elements dynamically.

# **Structural component example**

# **app.html**



```
<button (click)="toggle()">Toggle Heading</button>
<h1 *ngIf="show">Heading Tag</h1>

```



# **app.ts**



```
import { Component, signal } from '@angular/core';
import { NgIf } from '@angular/common';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, NgIf],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App{
  show = true;

  toggle() {
    this.show = !this.show;
  }
}

```

 1. What is *ngFor?

*ngFor is a structural directive in Angular used to loop through collections (arrays/objects) and dynamically generate elements in the DOM.

# **app.html**



```
<h1>*ngFor directives in Angular</h1>
<ul>
  <li *ngFor="let x of student">
    {{ x }}
  </li>
</ul>

<h2>StudentData List</h2>
<ul>
  <li *ngFor="let students of studentData">
    {{ students.name }} - {{ students.age }} years - {{ students.email }}
  </li>
</ul>

```



# **app.ts**



```
import { Component } from '@angular/core';
import { NgFor } from '@angular/common';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, NgFor],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  student = ["ak", "jk", "bk", "jk", "sk"];

  studentData = [
    { name: 'Sam', age: 34, email: 'sam@test.com' },
    { name: 'Bruce', age: 23, email: 'bruce@test.com' },
    { name: 'Tony', age: 34, email: 'tony@test.com' },
    { name: 'Peter', age: 30, email: 'peter@test.com' },
    { name: 'Natasha', age: 27, email: 'natasha@test.com' },
    { name: 'Steve', age: 31, email: 'steve@test.com' },
    { name: 'Wanda', age: 29, email: 'wanda@test.com' },
    { name: 'Clint', age: 32, email: 'clint@test.com' },
    { name: 'Vision', age: 28, email: 'vision@test.com' },
    { name: 'Strange', age: 38, email: 'strange@test.com' }
  ];
}

```

# **app.html**



```
<h1>*ngIf directives in Angular</h1>

<h1 *ngIf="block === 1">Heading 1</h1>
<h1 *ngIf="block === 2">Heading 2</h1>
<h1 *ngIf="block === 3">Heading 3</h1>
<h1 *ngIf="block === 4">Heading 4</h1>

<button (click)="updateBlockVal()">Toggle Heading</button>

```



# **app.ts**



```
import { Component } from '@angular/core';
import { NgIf } from '@angular/common';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, NgIf],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  block = 1;

  updateBlockVal() {
    this.block = this.block < 4 ? this.block + 1 : 1;
  }
}

```
