# **1. What is Two Way Binding?**
Two-way binding allows data to flow both from the component to the template and from the template to the component. It means changes in the UI reflect in the component and vice versa.

Its a combination of property binding and event binding.

Uses the syntax: [(ngModel)]

# **2. Make Input Field for Two Way Binding**
An example of an input field using two-way binding:



```
<input [(ngModel)]="username" placeholder="Enter username">
<p>Hello, {{ username }}</p>

```



# **3. Import FormsModule**

To use ngModel, you must import FormsModule in your Angular module:



```
import { FormsModule } from '@angular/forms';

@NgModule({
  imports: [FormsModule],
})
export class AppModule { }

```



# **4. Apply ngModel**
ngModel is the directive used to implement two-way binding:



```
<input [(ngModel)]="userData">

```



# **app.html**



```
<h1>Two way binding</h1>
<input type="text" [(ngModel)]="name" placeholder="Enter name">
<h1>{{name}}</h1>

```



# **app.ts**



```
import { Component, signal } from '@angular/core';
import { FormsModule } from '@angular/forms';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, FormsModule],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  name = "Ajay";
}

```

# **TO-DO-LIST**

# **app.html**



```
<h1>To-Do List</h1>

<input [(ngModel)]="task" type="text" placeholder="enter task name">
<button (click)="addTask()">Add Task</button>

@for (task of taskList; track task.id) {
  <ul class="task-list">
    <li>{{ task.id }}</li>
    <li>{{ task.name }}</li>
    <li><button (click)="deleteTask(task.id)">Delete</button></li>
  </ul>
}

```



# **app.ts**



```
import { Component, signal } from '@angular/core';
import { FormsModule } from '@angular/forms';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, FormsModule],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  task = "meeting at 10";
  taskList: { id: number, name: string }[] = [];

  addTask() {
    this.taskList.push({
      id: this.taskList.length + 1,
      name: this.task
    });
    this.task = "";
  }

  deleteTask(id: number) {
    this.taskList = this.taskList.filter((item) => item.id != id);
  }
}


```

# ** 1. What is Dynamic Styling?**
Dynamic styling in Angular means changing CSS styles or classes at runtime based on component state or user interactions.

It helps make the UI responsive to data or events.

You can apply it using:



```
[style.property] or [ngStyle] for inline styles

[class.className] or [ngClass] for CSS classes
```





image.png

# **app.html**



```
<!-- Commented out static style for reference -->
<!-- <h1 style="background-color: red;">Dynamic Style</h1> -->

<!-- Dynamic color and font size -->
<h1 [style.color]="color" [style.fontSize]="fontSizeNormal">Heading Tag</h1>

<!-- Conditional font size using ternary operator -->
<h1 [style.fontSize]="zoom ? fontSizeBig : fontSizeNormal">Heading tag 2</h1>

<!-- Button to toggle zoom -->
<button (click)="zoomToggle()">Zoom Toggle</button>

<!-- Font size with units (px) binding -->
<h1 [style.fontSize.px]="size">Heading tag 3</h1>

```



# **app.ts**



```
import { Component, signal } from '@angular/core';
import { FormsModule } from '@angular/forms';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, FormsModule],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  color = "green";                     // string
  fontSizeNormal = "30px";            // string with unit
  fontSizeBig = "60px";               // string with unit
  size = 30;                           // number only (used with .px binding)
  zoom = false;                        // boolean

  zoomToggle() {
    this.zoom = !this.zoom;
    this.size = this.zoom ? 60 : 30;   // update size value accordingly
  }
}

```
