```
<a [routerLink]="['/profile', 'Ajay']">Go to Profile</a>

```

```
import { Component } from '@angular/core';
import { Router, RouterLink } from '@angular/router';

@Component({
  selector: 'app-home',
  standalone: true,
  imports: [RouterLink],
  templateUrl: './home.html',
  styleUrls: ['./home.css']
})
export class Home {
  constructor(private router: Router) {}

  goToProfile() {
    this.router.navigate(['profile'], {
      queryParams: { name: 'Ajay Karre' }
    });
  }
}

```

```
import { Routes } from '@angular/router';
import { Login } from './login/login';
import { About } from './about/about';
import { Contact } from './contact/contact';
import { Home } from './home/home';
import { PageNotFound } from './page-not-found/page-not-found';
import { User } from './user/user';
import { Profile } from './profile/profile';

export const routes: Routes = [
  { path: 'login', component: Login },
  { path: 'about', component: About },
  { path: 'contact', component: Contact },
  { path: '', component: Home },
  { path: 'profile', component: Profile }, // ✅ Fixed: added comma
  { path: '**', component: PageNotFound }
];

```

```
import { Component } from '@angular/core';
import { ActivatedRoute } from '@angular/router';

@Component({
  selector: 'app-profile',
  standalone: true,
  imports: [],
  templateUrl: './profile.html',
  styleUrls: ['./profile.css']
})
export class Profile {
  userName: string | null = '';

  constructor(private route: ActivatedRoute) {}

  ngOnInit() {
    this.userName = this.route.snapshot.queryParamMap.get('name');  // ✅ fixed here
  }
}

```

# **🔹 When to use Reactive Forms**
Use them when you need more control over form logic, validation, and reactive programming patterns.

Ideal for dynamic forms, complex validations, and unit testing.



```
🔹 Make HTML Form
Create a regular HTML structure using <input>, <form>, <button>, etc.

Bind each field using [formControl] or [formGroup].

```



image.png

# **app.html**



```
<h1>Basic of Reactive Forms</h1>

<input type="text" placeholder="Enter name" [formControl]="name" />
<br />

<input type="text" placeholder="Enter password" [formControl]="pswd" />
<br />

<input type="text" placeholder="Enter email" [formControl]="email" />
<br /><br />

<button (click)="getValue()">Get Info</button>
<button (click)="setValue()">Set Info</button>

<h1>{{ name.value }}</h1>
<h1>{{ pswd.value }}</h1>
<h1>{{ email.value }}</h1>

```



# **app.css**



```
h1 {
  color: red;
  background-color: orange;
}

```



# **app.ts**



```
import { Component } from '@angular/core';
import { RouterOutlet } from '@angular/router';
import { FormControl, ReactiveFormsModule } from '@angular/forms';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, ReactiveFormsModule],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  name = new FormControl('bk');
  pswd = new FormControl('@123');
  email = new FormControl('karreajay36@gmail.com');

  getValue() {
    console.log(this.name.value);
    console.log(this.pswd.value);
    console.log(this.email.value);
  }

  setValue() {
    this.name.setValue('ak');
    this.pswd.setValue('123@');
    this.email.setValue('akmr227@gmail.com');
  }
}

```

# **✅ What is Form Grouping in Angular?**
Form Grouping in Angular means organizing multiple form controls (FormControl) into a single object called a FormGroup.

# **🔍 Why use FormGroup?**
To group related form fields together (e.g., name, email, password).

Makes it easier to validate, track, and manage multiple controls as a single unit.

Useful when building reactive forms.





```
profileForm = new FormGroup({
  name: new FormControl(''),
  email: new FormControl(''),
  password: new FormControl('')
});

```



# **🎯 Summary:**

FormControl → single input.

FormGroup → collection of inputs.

Together, they help build dynamic and scalable forms.

# **app.html**



```
<h1>Form Grouping in Reactive Forms</h1>

<form [formGroup]="profileForm" (ngSubmit)="submitData()">
  <input type="text" placeholder="Enter name" formControlName="name" />
  <br /><br />
  <input type="text" placeholder="Enter password" formControlName="password" />
  <br /><br />
  <input type="text" placeholder="Enter email" formControlName="email" />
  <br /><br />
  <button type="submit">Submit</button>
  <button type="button" (click)="setValues()">SetValues </button>
</form>

<h3>{{this.profileForm.value.name}}</h3>



```



# **app.ts**



```
import { Component } from '@angular/core';
import { FormControl, FormGroup, ReactiveFormsModule } from '@angular/forms';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [ReactiveFormsModule],
  templateUrl: './app.html',
  styleUrl: './app.css'
})
export class App {
  profileForm = new FormGroup({
    name: new FormControl(''),
    password: new FormControl(''),
    email: new FormControl('')
  });

  submitData() {
    console.log(this.profileForm.value);
  }
  setValues(){
    this.profileForm.setValue({
      name: 'John Doe',
      password: 'password123',
      email: 'johndoe@example.com'
    });
    
  }
}

```
