**In Angular, services are reusable pieces of logic or code that perform specific tasks and can be shared across multiple components. They help in separating business logic from the UI code (components), which makes your application more organized and maintainable.**

**✅ Key points:**

Services are usually used to fetch data, communicate with APIs, or share data between components.

They are created using @Injectable() decorator and are injected into components using Dependency Injection (DI).

# **app.ts**



```
import { Component } from '@angular/core';
import { ProductService } from './services/product'; // ✅ Import the service

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [],
  templateUrl: './app.html',
  styleUrl: './app.css'
})
export class App {
  productData: {
    name: string;
    branch: string;
    price: string;
  }[] | undefined;

  constructor(private productService: ProductService) {}

  getProductData() {
    this.productData = this.productService.getProductData();
    console.log(this.productData);
  }
}

```



# **app.html**



```
<h1>Service</h1>

<button (click)="getProductData()">Load Product Data</button>

@for (item of productData; track item) {
  <h1>{{item.name}},{{item.name}},{{item.name}}</h1>
}

```

# **app.ts**



```
import { Component, OnInit } from '@angular/core';
import { ProductService } from './services/product';
import { HttpClientModule } from '@angular/common/http';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [HttpClientModule], // ✅ Ensure HttpClientModule is included
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  productList: any;

  constructor(private productService: ProductService) {}

  ngOnInit() {
    this.productService.getProductList().subscribe((data: any) => {
      console.log(data);
      this.productList = data.products;
    });
  }
}

```



# **app.html**



```
<h1>API call services</h1>

@for (product of productList; track product.id) {
  <ul>
    <li>{{ product.title }}</li>
    <li>{{ product.price }}</li>
    <li>{{ product.rating }}</li>
    <li>{{ product.category }}</li>
  </ul>
}

```



# **product.ts**



```
import { Injectable } from '@angular/core';
import { HttpClient } from '@angular/common/http';

@Injectable({
  providedIn: 'root'
})
export class ProductService {
  constructor(private http: HttpClient) {}

  getProductList() {
    const url = "https://dummyjson.com/products";
    return this.http.get(url);
  }
}

```
