# **what is mean by Custom Componen?**

In Angular, a Custom Component refers to a user-defined reusable UI element created using Angular's @Component decorator. It allows you to encapsulate HTML, CSS, and logic into a self-contained unit that can be used like a built-in HTML tag.

# Note:In Angular (and TypeScript classes), if you want to call another function from within an existing method, use the this keyword to access the function.



```
export class MyComponent {
  welcome() {
    console.log("Welcome to Angular!");
  }

  ngOnInit() {
    this.welcome(); // calling another method using 'this'
  }
}

```

# **🔹 1. Why we need to define type**

Helps catch errors at compile time

Improves code readability and IntelliSense support



# **🔹 2. Define type for property**



```
export class AppComponent {
  title: string = 'My Angular App';
  age: number = 25;
}

```



# **🔹 3. Define multiple types for a property**



```
export class AppComponent {
  status: string | boolean = true; // can be either string or boolean
}

```



# **🔹 4. Define type for function parameters**


```
export class AppComponent {
  greet(name: string, age: number): void {
    console.log(`Hello ${name}, Age: ${age}`);
  }
}

```



# **🔹 5. Call function with params on button click**



```
<button (click)="greetUser('Ajay', 22)">Greet</button>

```


```
greetUser(name: string, age: number) {
  alert(`Hello ${name}, Age: ${age}`);
}

```
