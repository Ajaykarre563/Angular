

# **app.ts**



```
import { Component, signal, computed } from '@angular/core';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  x = signal(20);
  y = signal(30);
  data = computed(() => this.x() + this.y());

  updateValue() {
    this.x.set(100);
  }

  updateXValue() {
    this.x.set(1000);
  }
}

```

# ** What is Effect in Angular?**

An effect() is a reactive function that automatically runs when any of the signals it depends on are updated. Its useful for side-effects (logging, DOM updates, API calls, etc.).

# **app.html**



```
<h1>Effect in Angular</h1>
<h1>{{ userName() }}</h1>

<button (click)="userName.set('Sam')">Update userName</button>
@if (displayHeading) {
  <h1 style="background-color: green; color: white; padding: 10px;">
    Main Heading
  </h1>
}

```



# **app.ts**



```
import { Component, effect, signal, computed } from '@angular/core';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  userName = signal('anil');
  count = signal(0);
  displayHeading = true;

  constructor() {
    effect(() => {
      if (this.count() == 2) {
        this.displayHeading = false
        setTimeout(() => {
          this.displayHeading = false
        }, 7000);
      } else {
        this.displayHeading = false
      }
    });
  }
}
```

**Contextual variables are special variables that are automatically available within a specific context, such as inside a loop, conditional, or template block. They give you information about the current state of the operation.**

# **Contextual Variables in a For Loop:**
**$count**

Represents the current iteration number (starting from 1).

**$index**

Represents the current index of the loop (starting from 0).

**$last**

A boolean indicating whether the current iteration is the last one.

**$first**

A boolean indicating whether the current iteration is the first one.

**$odd & $even**

Booleans used to check if the current iteration is odd or even.
