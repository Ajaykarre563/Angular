# **app.html**



```
<h1>Counter App</h1>

<div style="text-align: center;">
  <h3 class="count">{{ count }}</h3>

  <div>
    <button (click)="handleCounter('minus')" class="btn bg-red">Decrement</button>
    <button (click)="handleCounter('reset')" class="btn bg-black">Reset</button>
    <button (click)="handleCounter('plus')" class="btn bg-green">Increment</button>
  </div>
</div>

```



# **app.css**



```
/* General button style */
.btn {
  width: 80px;
  height: 80px;
  border-radius: 50%;
  border: 1px solid #000;
  margin: 10px;
  font-weight: bold;
  color: white;
  cursor: pointer;
  font-size: 14px;
}

/* Specific background colors */
.bg-red {
  background-color: red;
}

.bg-black {
  background-color: black;
}

.bg-green {
  background-color: green;
}

/* Center the content */
.center {
  text-align: center;
  margin-top: 40px;
}

/* Style for the counter display */
.count {
  background-color: grey;
  display: inline-block;
  padding: 10px 30px;
  border-radius: 10px;
  font-size: 24px;
  font-weight: bold;
  color: black;
  margin-bottom: 20px;
}
```



# **app.ts**



```
import { Component } from '@angular/core';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  count = 0;

  handleCounter(action: string) {
    if (action === 'plus') {
      this.count++;
    } else if (action === 'minus') {
      this.count--;
    } else if (action === 'reset') {
      this.count = 0;
    }
  }
}

```



# **app.ts**



```
import { Component } from '@angular/core';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  username = "";

  // Called when user types in the input field
  getUserName(event: Event) {
    this.username = (event.target as HTMLInputElement).value;
  }

  // Called when "Set User name" button is clicked
  setUserName() {
    this.username = "Peter";
  }

  // Called when using template reference variable
  getUserNameWithTemplate(input: HTMLInputElement) {
    this.username = input.value;
    console.log(input.value);
  }
}

```

# **Style Option and Rules**

image.png

# **1.Component Style**
Each Angular component can have its own CSS file.

Defined using the styleUrls property in the @Component decorator.



```
@Component({
  selector: 'app-example',
  templateUrl: './example.component.html',
  styleUrls: ['./example.component.css']
})

```



# **2. Global Style**
Styles that are applied application-wide.

Usually defined in styles.css or styles.scss at the root level (defined in angular.json).



image.png

# **3. Make New File for Global Style**
You can create additional global CSS/SCSS files and import them into angular.json under the styles array.



```
"styles": [
  "src/styles.css",
  "src/custom-global-styles.css"
]

```



# **🔹 4. Internal Style**
CSS defined directly within the component HTML using <style> tags.



```
<style>
  h1 { color: red; }
</style>

```



# **🔹 5. Style in Component.ts File**
You can define styles inline using the styles array in the @Component decorator.



```
@Component({
  selector: 'app-inline-style',
  template: `<h1>Hello</h1>`,
  styles: [`h1 { color: green; }`]
})

```



# **🔹 6. Load Multiple CSS Files in Component**
In the styleUrls array, you can include multiple CSS files.



```
@Component({
  selector: 'app-example',
  templateUrl: './example.component.html',
  styleUrls: ['./style1.css', './style2.css']
})

```
