# **app.html**



```
<h1>*ngSwitch directives in Angular</h1>

<button (click)="changeColor('red')">Red color</button>
<button (click)="changeColor('green')">Green color</button>
<button (click)="changeColor('blue')">Blue color</button>
<button (click)="changeColor('orange')">Orange color</button>
<button (click)="changeColor('black')">Black color</button>
<button (click)="changeColor('yellow')">Yellow color</button>

<div [ngSwitch]="color">
  <h1 *ngSwitchCase="'red'" style="background-color: red;">Red</h1>
  <h1 *ngSwitchCase="'green'" style="background-color: green;">Green</h1>
  <h1 *ngSwitchCase="'blue'" style="background-color: blue;">Blue</h1>
  <h1 *ngSwitchCase="'orange'" style="background-color: orange;">Orange</h1>
  <h1 *ngSwitchCase="'black'" style="background-color: black; color: white;">Black</h1>
  <h1 *ngSwitchCase="'yellow'" style="background-color: yellow;">Yellow</h1>
  <h1 *ngSwitchDefault>No color Match</h1>
</div>

```



# **app.ts**



```
import { Component } from '@angular/core';
import { NgSwitch, NgSwitchCase, NgSwitchDefault } from '@angular/common';
import { RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, NgSwitch, NgSwitchCase, NgSwitchDefault],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})
export class App {
  color = ''; //  Declare color variable

  changeColor(color: string) {
    this.color = color;
  }
}

```

# ** Basic Routing in Angular**
# **What is Routing**

Routing in Angular allows navigation between different components (pages) in a single-page application (SPA) without refreshing the whole page.

# **Make Some Components for Pages**

Create different components (e.g., Home, About, Contact) using:

# ** What is router-outlet tag?**
router-outlet tag is a special Angular tag that tells Angular where to display the component for the current route.

Think of it like a screen or placeholder  Angular looks at the current URL and loads the matching component inside that placeholder.

# **app.html**



```
<h1>Basic Routing</h1>

<ul>
  <li><a routerLink="login">Login</a></li>
  <li><a routerLink="about">About</a></li>
  <li><a routerLink="contact">Contact</a></li>
</ul>

<router-outlet></router-outlet>

```



# **app.routes.ts**



```
import { Routes } from '@angular/router';
import { Login } from './login/login';
import { About } from './about/about';
import { Contact } from './contact/contact';

export const routes: Routes = [
  { path: 'login', component: Login },
  { path: 'about', component: About },
  { path: 'contact', component: Contact },
];

```



# **app.ts**



```
// app.component.ts
import { Component } from '@angular/core';
import { RouterLink, RouterOutlet } from '@angular/router';

@Component({
  selector: 'app-root',
  standalone: true,
  imports: [RouterOutlet, RouterLink],
  templateUrl: './app.html',
  styleUrls: ['./app.css']
})

export class App {}

```
