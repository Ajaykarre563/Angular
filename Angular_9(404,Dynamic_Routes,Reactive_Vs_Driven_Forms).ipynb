```
import { Routes } from '@angular/router';
import { Login } from './login/login';
import { About } from './about/about';
import { Contact } from './contact/contact';
import { Home } from './home/home';
import { PageNotFound } from './page-not-found/page-not-found'; // ✅ Fixed name

export const routes: Routes = [
  { path: 'login', component: Login },
  { path: 'about', component: About },
  { path: 'contact', component: Contact },
  { path: '', component: Home },
  { path: '**', component: PageNotFound } // ✅ Catch-all route
];

```





```
✅ Angular Wildcard Route (**) – Key Note:

The ** path in Angular routing is a wildcard route, used to catch any URLs not matched by previous route definitions.

It should always be placed last in the Routes array, because Angular checks routes top to bottom, and matches the first one it finds.

Typically used to show a Page Not Found component.
```

```
import { Routes } from '@angular/router';
import { Login } from './login/login';
import { About } from './about/about';
import { Contact } from './contact/contact';
import { Home } from './home/home';
import { PageNotFound } from './page-not-found/page-not-found';
import { User } from './user/user'; // ✅ Don't forget to import this

export const routes: Routes = [
  { path: 'login', component: Login },
  { path: 'about', component: About },
  { path: 'contact', component: Contact },
  { path: '', component: Home },
  { path: 'user/:id/:name', component: User }, // ✅ Added missing comma
  { path: '**', component: PageNotFound }      // ✅ Wildcard route for unmatched paths
];

```



# **user.ts**



```
import { Component, OnInit } from '@angular/core';
import { ActivatedRoute, Params } from '@angular/router';

@Component({
  selector: 'app-user',
  templateUrl: './user.html',
  styleUrls: ['./user.css']  // ✅ Fixed typo: styleUrl ➝ styleUrls
})
export class User {
  name: string | null = '';
  id: string | null = '';

  constructor(private route: ActivatedRoute) {}

  ngOnInit() {
    this.route.params.subscribe((params) => {
      this.name = params['name'];
      this.id = params['id'];
    });
  }
}
```

# 🧠 **Reactive Forms**

Best for complex forms.

You write the form logic in the TypeScript code.

More control, testable, and scalable.

Great when forms are a big part of your app.

Example: Used in apps with dynamic forms, validation rules, or reusable components.

# **📝 Template-driven Forms**

Best for simple forms.

You write most of the logic in the HTML template using Angular directives.

Easy to set up, but harder to manage as it gets complex.

Example: A simple contact form or newsletter sign-up.
